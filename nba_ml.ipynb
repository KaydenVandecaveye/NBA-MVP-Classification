```python
#Actual Precision Function
def find_ap(combination):
    actual = combination.sort_values(["Share"], ascending = False).head(5)
    predicted = combination.sort_values(["Predictions"], ascending = False)
    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
find_ap(combination)
```

```python
years = list(range(1980,2025))
aps = []
all_preds = []
for year in years[5:]:
    train = stats[stats["Year"] < year]
    test = stats[stats["Year"] == year]
    reg.fit(train[preds], train[target])
    reg_predicts = reg.predict(test[preds])
    reg_predicts = pd.DataFrame(reg_predicts, columns = ["Predictions"], index = test.index)
    combination = pd.concat([test[["Player","Share"]], reg_predicts], axis = 1)
    all_preds.append(combination)
    aps.append(find_ap(combination))
    
```

```python
#Mean AP
sum(aps) / len(aps)
```

```python
def add_ranks(combination):
    combination = combination.sort_values(["Share"], ascending = False)
    combination["Rank"] = list(range(1,combination.shape[0] + 1))
    combination = combination.sort_values(["Predictions"], ascending = False)
    combination["Predicted Rank"] = list(range(1,combination.shape[0] + 1))
    combination["Difference"] = combination["Rank"] - combination["Predicted Rank"]
    return combination
```

```python
ranking = add_ranks(all_preds[1])
ranking[ranking["Rank"] < 10].sort_values(["Difference"], ascending = False)
```

```python
def backtest(stats, model, year, preds):
    aps = []
    all_preds = []
    for year in years[5:]:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[preds], train["Share"])
        predicts = model.predict(test[preds])
        predicts = pd.DataFrame(predicts, columns = ["Predictions"], index = test.index)
        combination = pd.concat([test[["Player","Share"]], predicts], axis = 1)
        combination = add_ranks(combination)
        all_preds.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_preds)
```
